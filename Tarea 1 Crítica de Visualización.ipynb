### PRIMER GRAFICO

En la siguiente visualizacíon se muestra los miedos mas inflamados por los medios desde el año 2000 hasta el 2018

 - ***¿QUÉ?, ¿POR QUÉ? Y ¿CÓMO?***
     - ¿WHAT?: Tabla, Atributos temporales y categóricos
     - ¿WHY?: Presentar, Descubrir, Disfrutar, Identificar y Distribución
     - ¿HOW?: Map, Navegar, Filtrar y Sobreponer

 - ***MARCAS Y CANALES USADOS***
     - MARCA: Área
     - CANAL: Color que representa cada una de las calamidades causas en el tiempo y eje horizontal que representan los años.

- ***¿CUMPLE O NO CUMPLE ALGUNAS LAS REGLAS GENERALES***
    - Cumple el principio de overview, zoom y filtro, detalles a petición dado que permite interactuar en la gráfica haciendo zoom y filtros por calamidad, además al dar click en cada destrucción aparece un cuadro con el detalle con la calamidad ocurrida

- ***MEJORAS POSIBLES***
    - considero que no es necesario realizar una mejora a la visualización ya que como se encuentra aporta mucha información interactiva y mucho detalle al ojo que permite entender de forma rápida lo que se quiere reflejar

### SEGUNDO GRAFICO

En la siguiente visualizacíon se muestra el desglose escena por escena de las "Historias Reales" de Hollywood

- ***¿QUÉ?, ¿POR QUÉ? Y ¿CÓMO?***
     - ¿WHAT?: Tabla -> Atributos -> Secuenciales
     - ¿WHY?: Presentar, Descubrir, Identificar y Depencia dado que cada escena depende de la anterior
     - ¿HOW?: Map, Color, Navegar y Filtrar

 - ***MARCAS Y CANALES USADOS***
     - MARCA: Área
     - CANAL: Color que representa en cada una de las escenas de las películas si es Cierto, Verdadero, Falso o Desconocido en la realidad y eje horizontal que representan la duración de cada película

- ***¿CUMPLE O NO CUMPLE ALGUNAS LAS REGLAS GENERALES***
    - Esta visualización también cumple el principio de overview, zoom y filtro, detalles que permite interactuar en la gráfica haciendo zoom y filtros por escena, además al dar click en cada color aparece un cuadro con el detalle de lo que pasó en la película y lo relaciona con la realidad

- ***MEJORAS POSIBLES***
    - considero que no es necesario realizar una mejora a la visualización ya que como se encuentra aporta mucha información y nos permite entender de forma rápida lo que se quiere reflejar 

***FUENTE*** https://informationisbeautiful.net/visualizations/based-on-a-true-true-story/
